Question 1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

Answer-

Type I Error (False Positive)

A Type I error occurs when we reject a true null hypothesis.

You conclude there is an effect, when in reality there is no effect.

Think of it as a false alarm.

Probability of Type I Error = Î± (alpha)

Common values: 0.05, 0.01

If Î± = 0.05, you accept a 5% risk of falsely rejecting a true null hypothesis.

Example

A medical test incorrectly indicates a patient has a disease when they actually do not.

Type II Error (False Negative)

A Type II error occurs when you fail to reject a false null hypothesis.

You conclude there is no effect, when an effect actually exists.

Think of it as missing something real.

Probability of Type II Error = Î² (beta)

Power of a test = 1 âˆ’Î²

Higher power = lower chance of Type II error.

Example

A medical test fails to detect a disease the patient actually has.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

Answer-
What Is the P-Value?

The P-value is the probability of obtaining a sample result at least as extreme as the one observed, assuming the null hypothesis is true.

In simpler terms:

P-value = How surprising your data is if the null hypothesis were true.

How to Interpret the P-Value
Small P-value (typically â‰¤ 0.05)

The observed result would be very unlikely if the null hypothesis were true.

Therefore, you have enough evidence to reject the null hypothesis.

Interpretation:
The data contradicts Hâ‚€; there is evidence of a real effect.

Alternative hypothesis (Hâ‚):
At least one group mean is different (but ANOVA does not specify which one).

ANOVA tests whether the variance between groups is significantly larger than the variance within groups.

How ANOVA Extends Hypothesis Testing to More Than Two Groups

A t-test only compares two means.
ANOVA allows comparison of the means of three or more groups at once without increasing Type I error.

Why not use many t-tests?

Running multiple t-tests increases the chance of false positives (Type I errors).

Example:

3 groups â†’ 3 t-tests

4 groups â†’ 6 t-tests

5 groups â†’ 10 t-tests
â€¦ and so on.

ANOVA avoids this by performing one overall test that checks for any difference among all group means.

If ANOVA is significant, post-hoc tests (like Tukeyâ€™s HSD) can determine which groups differ.

Assumptions of ANOVA

ANOVA relies on three key assumptions:

1. Independence of Observations

Each data point must be independent.

No repeated measures or paired data unless using special forms (e.g., repeated-measures ANOVA).

2. Normality

The data in each group should be approximately normally distributed.

ANOVA is somewhat robust to this assumption for large samples.

3. Homogeneity of Variances (Equal Variances)

All groups should have similar variances.

Tested using Leveneâ€™s or Bartlettâ€™s test.

If variances are not equal, alternatives like Welchâ€™s ANOVA can be used.

```python
Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset

Answer-

import numpy as np
from scipy import stats

# -----------------------------
# Given Data
# -----------------------------
sample = [12, 15, 14, 10, 13, 14, 16, 15, 14, 13]  # sample data
population_mean = 13                                # hypothesized population mean
population_std = 2                                  # population standard deviation (known)
alpha = 0.05                                        # significance level

# -----------------------------
# Z-test Calculation
# -----------------------------
sample_mean = np.mean(sample)
sample_size = len(sample)
standard_error = population_std / np.sqrt(sample_size)

# Z-statistic
z_stat = (sample_mean - population_mean) / standard_error

# p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# -----------------------------
# Print Results
# -----------------------------
print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

# -----------------------------
# Interpretation
# -----------------------------
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference between the sample mean and the population mean.")
```

```python
Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.

Answer-

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Simulate Data
# -----------------------------
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples

data = np.random.binomial(n, p, size)

# -----------------------------
# Plot Histogram
# -----------------------------
plt.hist(data, bins=range(0, n+2), edgecolor='black', alpha=0.7)
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.xticks(range(0, n+1))

plt.show()
```

```python
Question 8: Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

Answer-

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Step 1: Generate non-normal data (Exponential Distribution)
# -----------------------------
np.random.seed(42)

population = np.random.exponential(scale=1, size=100000)  # highly skewed distribution

# -----------------------------
# Step 2: Draw many samples and compute sample means
# -----------------------------
sample_size = 50       # size of each sample
num_samples = 1000     # total samples drawn

sample_means = []

for _ in range(num_samples):
    sample = np.random.exponential(scale=1, size=sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# -----------------------------
# Step 3: Plot distributions
# -----------------------------
plt.figure(figsize=(14, 6))

# Plot 1: Original Non-Normal Distribution
plt.subplot(1, 2, 1)
sns.histplot(population, bins=40, kde=True, color='skyblue')
plt.title("Original Distribution (Exponential - Non Normal)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot 2: Sampling Distribution of Sample Means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=40, kde=True, color='salmon')
plt.title("Sampling Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
Question 9: Write a Python function to calculate and visualize the confidence interval
for a sample mean.

Answer-

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for the sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default = 0.95)
    """
    data = np.array(data)
    mean = np.mean(data)
    n = len(data)
    std = np.std(data, ddof=1)

    # Standard error
    se = std / np.sqrt(n)

    # t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Confidence interval
    lower = mean - t_crit * se
    upper = mean + t_crit * se

    print(f"Sample Mean: {mean:.3f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.3f}, {upper:.3f})")

    # -----------------------------
    # Visualization
    # -----------------------------
    plt.figure(figsize=(8, 4))
    plt.errorbar(x=mean, y=0, xerr=[[mean - lower], [upper - mean]],
                 fmt='o', color='red', ecolor='black', capsize=8)

    plt.scatter(data, np.zeros(len(data)), alpha=0.5, label="Data Points")

    plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
    plt.axvline(lower, color='blue', linestyle='--', label="Lower CI")
    plt.axvline(upper, color='blue', linestyle='--', label="Upper CI")

    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.xlabel("Value")
    plt.yticks([])  # hide y-axis
    plt.legend()
    plt.show()

    return lower, upper

# Example Usage
sample_data = [12, 15, 14, 13, 16, 14, 15, 17, 18, 14]
plot_confidence_interval(sample_data, confidence=0.95)
```

```python
Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

Answer-

import numpy as np
from scipy.stats import chisquare

# -----------------------------
# Step 1: Define observed and expected frequencies
# -----------------------------
# Example: Observed counts of 6-sided die rolls
observed = np.array([18, 22, 20, 15, 17, 8])

# Expected counts if the die is fair
total = observed.sum()
expected = np.array([total/6]*6)

# -----------------------------
# Step 2: Perform Chi-square goodness-of-fit test
# -----------------------------
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi2_stat)
print("P-Value:", p_value)

# -----------------------------
# Step 3: Interpret the result
# -----------------------------
alpha = 0.05  # significance level

if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: Observed distribution significantly differs from expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: Observed distribution is consistent with expected distribution.")
```
